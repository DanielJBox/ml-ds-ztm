## 0. An end-to-end Scikit-Learn workflow

```python
# 1. Get the data ready
import pandas as pd
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# Create X (features matrix)
X = heart_disease.drop("target", axis=1)
X
```

```python
# Create y (labels)
y = heart_disease["target"]
y
```

```python
# 2. Choose the right model (estimator/algorithm) and hyperparameters for our problem
# Hyperparameters: Like dials on a model that we can choose to make it better/worse
# Our problem is classification because we want to put each person, each row in the 
# class of 'has heart disease' or 'does not have heart disease'.
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf.fit(X_train, y_train)
```

```python
# make a prediction
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
