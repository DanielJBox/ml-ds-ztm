## Two options for subplots

```python
# Subplots option 1
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

## matplotlib OO explicit method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
# Pyplot method - Not recommended (Ok for quick plots)
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
plt.show()
```

```python
# Matplotlib OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
ax.set_xlim([45, 100])
plt.show()
```

```python
over_50.target.values
```

```python
# Matplotlib OO method (recommended)
# fig, ax = plt.subplots(figsize=(10, 6))
fig, ((ax_age, ax_thal), (ax_sex, ax_spare)) = plt.subplots(nrows=2,
                                                           ncols=2,
                                                           figsize=(10, 10))

fig.suptitle("Heart Disease and Cholesterol Levels", fontsize="16", fontweight="bold")

# Sub Plot of age and cholesterol
# Plot the data
scatter = ax_age.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax_age.set(title="By Age",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a horizontal line
mean_chol = over_50["chol"].mean()
mean_line = ax_age.axhline(mean_chol,
                           linestyle='--',
                           color='green',
                           label="Mean Cholesterol")

# Add a legend
# Get scatter legend elements
handles, labels = scatter.legend_elements()

# Modify labels to be 'No' and 'Yes' instead of 0 and 1
labels = ["0 - No", "1 - Yes"]

scatter_legend = ax_age.legend(handles, labels, title="High Chol? (Target)", loc="upper right")
ax_age.add_artist(scatter_legend)
ax_age.legend(handles=[mean_line], loc="lower right")

# Sub Plot of thalach (max heart rate) and cholesterol
# Plot the data
scatter = ax_thal.scatter(x=over_50["thalach"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax_thal.set(title="By Thalach (max heart rate)",
       xlabel="Thalach",
       ylabel="Cholesterol");

# Add a horizontal line
mean_line = ax_thal.axhline(mean_chol,
                           linestyle='--',
                           color='green',
                           label="Mean Cholesterol")

# Add a legendy
# Get scatter legend elements
handles, labels = scatter.legend_elements()

# Modify labels to be 'No' and 'Yes' instead of 0 and 1
labels = ["0 - No", "1 - Yes"]

scatter_legend = ax_thal.legend(handles, labels, title="High Chol? (Target)", loc="upper right")
ax_thal.add_artist(scatter_legend)
ax_thal.legend(handles=[mean_line], loc="lower right")

# Sub Plot of sex and cholesterol
chol_by_sex = over_50.groupby("sex")["chol"].mean()

bar = ax_sex.bar(x=[0, 1],
                 height=chol_by_sex,
                 color=['blue', 'pink'])

ax_sex.set(title="By Sex",
           xlabel="Sex",
           ylabel="Cholesterol")

ax_sex.set_xticks([0, 1])
ax_sex.set_xticklabels(["Men", "Women"])

mean_line = ax_sex.axhline(mean_chol,
                           linestyle='--',
                           color='green',
                           label="Mean Cholesterol")

ax_sex.legend(handles=[mean_line],loc="lower right")

plt.show()
```
