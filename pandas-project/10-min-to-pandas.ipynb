## Merge

```python
# Concat
df5 = pd.DataFrame(np.random.randn(10, 4))
df5
```

```python
pieces = [df5[:3], df5[3:7], df5[7:]]
pieces
```

```python
pd.concat(pieces)
```

```python
# Join
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})
left
right
pd.merge(left, right, on="key")
```

```python
left = pd.DataFrame({"key": ["foo", "bar"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "bar"], "rval": [4, 5]})
pd.merge(left, right, on="key")
```

```python
# Grouping
df6 = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df6

```

```python
df6.groupby("A")[["C", "D"]].sum()
```

```python
df6.groupby(["A", "B"]).sum()
```

```python
# Reshaping
# Stack
arrays = [
   ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
   ["one", "two", "one", "two", "one", "two", "one", "two"],
]
index = pd.MultiIndex.from_arrays(arrays, names=["first", "second"])
df7 = pd.DataFrame(np.random.randn(8, 2), index = index, columns=["A", "B"])
df7
```

```python
stacked = df7.stack(future_stack=True)
stacked
```

```python
stacked.unstack()
```

```python
# Pivot tables
df8 = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 3,
        "B": ["A", "B", "C"] * 4,
        "C": ["foo", "foo", "foo", "bar", "bar", "bar"] * 2,
        "D": np.random.randn(12),
        "E": np.random.randn(12),
    }
)


df8
```

```python
pd.pivot_table(df8, values="D", index=["A", "B"], columns=["C"])
```

## Categoricals

```python
df9 = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
df9
```

```python
df9["grade"] = df9["raw_grade"].astype("category")
df9
```

```python
df9["grade"]
```

```python
new_categories = ["very good", "good", "very bad"]
df9["grade"] = df9["grade"].cat.rename_categories(new_categories)
df9
```

```python
df9["grade"] = df9["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"]
)
df9
```

```python
df9["grade"]
```

```python
df9.sort_values(by="grade")
```

```python
df9.groupby("grade", observed=False).size()
```
